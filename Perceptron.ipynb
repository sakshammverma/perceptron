```python
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt


X,y=make_classification(n_samples=100,n_features=2,n_informative=1,n_redundant=0,n_classes=2,n_clusters_per_class=1,random_state=41,hypercube=False,class_sep=10)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.show()
```

```python
def perceptron (X,y):
  X=np.insert(X,0,1,axis=1) #insert bias
  weights=np.ones(X.shape[1])  #+1 for bias
  lr=0.1


  for i in range(1000):
    random_value= np.random.randint(0,X.shape[0])
    predicted_value= np.dot(X[random_value],weights)


    if predicted_value > 0 :
      predicted_value=1
    else:
      predicted_value=0

    new_weight= weights + lr*(y[random_value]-predicted_value)*X[random_value]
    weights=new_weight

  return new_weight[0],new_weight[1:]
```

```python

intercept_, coef_ = perceptron(X,y)
```

```python

print(intercept_,coef_)
```

```python

#m=-A/B
m= -(coef_[0]/coef_[1])
b= -(intercept_/coef_[1])

```

```python
x_input= np.linspace(-3,3,100)
y= m*x_input + b

```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
